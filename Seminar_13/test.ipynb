```python
import os
import pandas as pd

# Папка с текстовыми файлами
folder_path = './Gogol/'

# Список файлов в папке
files = os.listdir(folder_path)

# Инициализация списков для столбцов "id" и "текст"
ids = []
texts = []

# Чтение текстов из файлов и заполнение списков
for file_name in files:
    file_path = os.path.join(folder_path, file_name)
    
    # Используйте любой способ чтения текста из файла, в зависимости от формата файлов
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Добавление в списки
    ids.append(file_name)
    texts.append(text)

# Создание DataFrame из списков
df2 = pd.DataFrame({'id': ids, 'text': texts})

# Сохранение DataFrame в CSV файл
df2.to_csv('D:/Develop/MachineLearning/Seminar_13/Gogol_1.csv', index=False)
df2.head()
```

```python
def Nltk(df, ind):
    df['new_col'] = df['text'].apply(lambda x: nlp(x))
    df['new_col'] = df['new_col'].apply(lambda x: [token.tag_ for token in x])

    tags = ['ADJ', 'ADP', 'ADV', 'AUX', 'CCONJ', 'DET', 'ENTITY', 'NOUN', 'PART', 'PRON', 'PUNCT', 'SCONJ', 'VERB']
    
    for i in tags:
        df[i] = df['new_col'].apply(lambda x: x.count(i))

    # Доля частей речи в тексте
    df['sum'] = df['new_col'].apply(lambda x: sum(x.count(i) for i in tags))

    for i in tags:
        df[i] = np.where(df['sum'] > 0, df['new_col'].apply(lambda x: x.count(i) / df['sum']), 0)

    for column in tags:
        df[column] = (df[column] - df[column].mean()) / (df[column].abs().max())

    # Добавление среднего положения частей речи
    for i in tags:
        df[f'{i}_position'] = df['new_col'].apply(lambda x: np.mean([idx for idx, tag in enumerate(x) if tag == i]))

    if ind == 1:    
        df['class'] = 1
    elif ind == 2:
        df['class'] = 0

    return df
```

```python
result_df1 = Nltk(df1, ind=1)
```

```python
result_df2 = Nltk(df2, ind=2)
```

```python
df = pd.concat([df1,df2], sort=False)
df.drop([ 'id', 'ENTITY', 'new_col', 'ENTITY_position'], axis=1, inplace=True)
df.head(10)
```

```python
df.drop([ 'text'], axis=1, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(df.drop(['class'], axis = 1), df['class'], test_size=0.2, random_state=42)
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

pres = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, y_pred)

print('prescision = {0}'.format(pres))
print('recall = {0}'.format(recall))
print('F1 = {0}'.format(f1))
print('Roc auc = {0}'.format(roc_auc))
```
